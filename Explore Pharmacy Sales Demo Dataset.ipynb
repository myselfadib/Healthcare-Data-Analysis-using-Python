### Medication Class Analysis

```python
df_ph.info()
```

```python
df_ph['Med_name'].value_counts()
```

```python
df_ph['Med_class'].value_counts()
```

```python
top_number = 20
```

```python
top_classes = df_ph['Med_class'].value_counts().nlargest(top_number)
top_classes 
```

```python
top_classes = df_ph['Med_class'].value_counts().nlargest(top_number).index
top_classes 
```

```python
## create a dataframe containing top 20 classes
df_top_med_classes = df_ph[df_ph['Med_class'].isin(top_classes)]
df_top_med_classes
```

```python
df_top_med_classes = df_top_med_classes.groupby('Med_class')['Quantity Sold'].sum().reset_index()
```

```python
df_top_med_classes
```

```python
df_top_med_classes = df_top_med_classes.sort_values(by='Quantity Sold')
```

```python
plt.barh(df_top_med_classes['Med_class'],df_top_med_classes['Quantity Sold'])
plt.title(f'Top {top_number} Medication by Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Medication Class')
plt.show()
```

# Demand Forcast

```python
!pip install prophet
```

```python
from prophet import Prophet
```

```python
df_ph
```

```python
df_prophet = df_ph[df_ph['Med_class'].str.contains('Anxiolytic')][['Date Sold', 'Quantity Sold']]
```

```python
df_prophet
```

```python
df_prophet = df_prophet.reset_index()
df_prophet
```

```python
df_prophet = df_prophet.rename(columns={'Date Sold':'ds','Quantity Sold':'y'})
df_prophet
```

```python
type(df_prophet['ds'][0])
```

```python
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])
```

```python
type(df_prophet['ds'][0])
```

```python
model = Prophet()
```

```python
model.fit(df_prophet)
```

```python
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)
```

```python
fig = model.plot(forecast)
plt.show()
```

```python
fig = model.plot_components(forecast)
plt.show()
```
