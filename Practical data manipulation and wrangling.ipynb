```python
df["Med_name"] = df[0].apply(lambda x: x.split(".", maxsplit=1)[1].split("-")[0])
df
```

```python
def split_and_assign(row):
    combined, med_class_2 = row[0].split("-", 1)
    return pd.Series({"combined": combined, "Med_class_2": med_class_2})

df = pd.concat([df, df.apply(split_and_assign, axis=1)], axis=1)
df
```

```python
df_med_2 = df[["Med_name","Med_class_2"]]
df_med_2
```

```python
df_med_2.rename(columns={"Med_class_2":"Medication_class"}, inplace=True)
df_med_2
```

```python
df_med_2 = df_med_2.drop([999])
df_med_2
```

```python
df_med_2.to_csv("Med_name_and_class.csv", index=False)
```
